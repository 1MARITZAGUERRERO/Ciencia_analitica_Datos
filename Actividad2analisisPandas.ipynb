8.	Ejecuta el siguiente código y comenta qué hace cada línea:

```python
datetime_df = air_df.ReadingDateTime.str.split(' ',expand=True) # El campo ReadingDataTime se esta separando en varias columnas separada por el caractes espacios
datetime_df.columns = ['Date','Time']  # se esta colocando un nombre a las  dos nuevas columnas
datetime_df  # se impreme el Dataframe
date_df = datetime_df.Date.str.split('/',expand=True) # El nuevo campo Date se esta separando en varias columnas  de acuerdo al /
date_df.columns = ['Day','Month','Year'] # se esta colocando nombre a las tres nuevas columnas
date_df # se impreme el Dataframe
air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year']) # Se unifican los tres dataframe, de esa unificacion se elimina los campos ReadingDateTime y Year, yo tambien el eliminaria el campo Date
air_df = air_df.set_index(['Month','Day','Time','Species'])  # Esta colocando como indice estas 4 variables del dataframe
air_df
```
